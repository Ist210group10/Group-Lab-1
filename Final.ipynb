Our Entity-Relationship Diagram will be used by the pet store in order to better track purchases and costs for the store. We have several entities in our ER diagram, including a separation of purchases for fish and pets, as well as their storage. Each entity has several attributes that can help the store better understand and organize customer trends.  The manager wanted fish and other pets to be kept separately, so we have an entity set for both Aquariums and Enclosures. Beyond that, we also separated Pet transactions and Fish Transactions. Each of those have a foreign key which relates back to the Aquarium and Enclosure entities. Cardinality is included to show the distinction between how many pets or fish have been sold in comparison to how many are left within the enclosures and aquariums, respectively. From the aquarium,150 have been sold and 50 remain. From the pet enclosures, 100 have been sold and 20 remain in the store. 
Food transactions are also separated from actual animal transactions, as was a request of the manager of the store. This gives the store a better idea as to whether they should adjust stock of different animal foods depending on sales data that can be collected through the database.
There is also a Customer entity, to track individual customer purchases to better cater to their needs as a consumer. When a customer makes a purchase, it can also be credited to their Loyalty account through their Loyalty ID. The loyalty program can be applied to a purchase of an animal or food for a store discount.
