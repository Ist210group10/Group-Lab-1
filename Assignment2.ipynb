We designed the ER diagram according to our Flight Cost idea from Assignment 1. We first started with adding the necessary tables, which include Flight, Passenger, Pilot, Airplane, and finally our real-time data is stored in Cost. Determing the flight cost for each pilot (in training) is our primary goal for this database. We wanted to shed some light on the cost of training for a pilot so that aspiring pilots know exactly what they're getting into. Gathering data from all of these tables and accumulating them into the cost table is how our real-time data will be calculated.

Inside of the tables, we added our necessary attributes. The Flight table has a Primary Key of FlightID because that FlightID is the unique ID for each flight. We also added data for each flight like flight time, flight distance, and number of passengers. Our foregin key in this table is our PilotID because we need to link each flight to a certain pilot. PilotID also has a 1 to (optional) many relationship between the pilot and passengers. The pilot could have up to many passengers.

The Passenger table is used to keep track of the amount of passengers per flight by our Primary Key, PassengerID, and the first name, last name, and our Foregin Key FlightID which ties each passenger to each flight. Keeping track of our passengers is important because that is one of our contraints for this database, since some pilots are only allowed to take so many passengers depending on their training and license.
The Airplane table is used to keep track of which plane each pilot is using, how much the rent for each plane is, and how much fuel it can carry. Our Primary Key is the AirplaneID to give each plane a unique ID number. We also have attributes of Model, Gas Tank Size, and Rent. Rent is our foreign key because it links to the pilot paying the rent for that particular plane. We also tie the Pilot to each plane as well. There is a 1 to 1 relationship between Pilot and PilotID to tie each pilot to each airplane.

The Pilot table contains each PilotID which gives each pilot a unique identifier. It also acts as a logbook for each pilot, which they need to use in training. The table holds the time flown, distance flown, instructor flown with, each flight ID, and real time cost for the pilot. The cost is our foreign key because it links back to the cost table and keeps the real-time data for the pilot's cost.

The final table is Cost. Cost uses all of the data stored in each table and uses that data to keep track of the real-time cost for the pilots (in training). The CostID keeps track of cost per flight for pilots. The cost table has attributes of cost of instructor, cost of fuel, and cost of rent. This table is used as our main table to give real-time data to aspiring pilots and pilots in training. The flight also has a 1 to (optional) many relationship between PilotID and Instructor. Because there is 1 pilot per 0 or many instructors.
