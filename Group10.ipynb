__Ride-sharing Service__



My idea of a ride-sharing service is using people to create self-employed taxis. People can call this ride through an app and bring the taxi to them, while the people driving are making a portion of the money we charge to host the service. Obviously, this idea is stolen from Uber. Although, I think it would be interesting to simulate some of this data and calculate what time of day or night would bring in the most revenue for the business or for the driver. Unlike Uber, this ride-sharing service would be able to call ahead for the driver, request certain drivers (male, female, foregin language speaker preferred, etc.) in order to maintain both rider and driver's quality of experience. The app would also allow set prices for the amount earned by the driver (higher than min. wage) for transparency. The business would also only take a set amount for their services provided. Finally, unlike Uber, the business would employ a set number of drivers at once, they can pick their own hours, but the service would have a minimum amount of drivers so they are never short and people are not waiting to get a ride. 

The data collected could serve as a tool to showcase when the business needs to have more or less drivers on, which also serves our goal of showcasing when the most money is earned.

The tables would include: 


1.   Time of ride (real-time)
2.   Set price earned by driver
2.   Set price of what service is owed (percantage?)
1.   How many drivers working at once
1.   Cost per distance (set cost per 1 mile, 2 miles, etc.)


A day to day interaction at my business would be drivers logging into work at a set time (9-5 or shorter, pick your own hours) and completing rides. Our database would work in the backgound to collect information about each ride (rating, time, price, etc.) and everything would work from a technology standpoint. 



Goal: To calculate what time of day/night brings in the most revenue. I would use "time of ride" and "money earned" tables to calculate this goal.

__Flight training__

---Civilian pilot training is a process that is not straight forward. Many different private schools offer this training process. Whether it is through a flying club, private airport, or a university or airline sponsored program, this process is different and tends to cost different amounts. This would be an online database to help young pilots not only understand the process and requirements of flight training but help them manage their progress and the cost. As a part of flight training, pilots will have to “log hours.” Logging hours is often done in a logbook which is usually pen and paper. This brings up some obvious issues. Pen and paper can easily get lost or ruined by a morning cup of coffee (something pilot are awfully fond of). By having a database that actively tracks cost for the pilot, they can have a better idea of what the total cost for the training will be. This idea can also be used once a pilot has graduated from training. Airplanes are expensive, so the majority of young pilots choose to rent planes in order to fly places or keep up their skills. This can also cost a person more than they expect. Planes can be rented by hours, days, weeks, or even months, but the cost of gas and flight distance can be calculated but can also be easily tracked by this database. The database is unique because there is no major company out there doing this. It would also provide transparency in cost making pilot training and flying accessible to a large population.

1.	Flight time (length and distance) 
1.	Price of fuel at take-off (price/gallon)
2.	Airplane used (model and year)
2.	Pilot (name and rank)
1. Cost &rarr; real-time input. This would track costs pre-flight, during flight, and after flight.


The goal of this idea is to get rid of the hidden costs and fear of getting into flying due to financial reasons. This will help to negate that fear factor and provide greater transparency over overall costs. This will be done by tracking realtime costs so a more universal cost can be discovered.

**Electric Scooter Rental Service**

The idea of an electric scooter rental service is people borrowing and riding electric scooters to get to their destinations conveniently. Basically, a business installs many electric scooters around places where there are a lot of people. For example, Universities, high schools, and downtown areas would be good places. And People can find the locations of unoccupied electric scooters around them by using GPS through an app. Once they find an electric scooter, they can activate it through QR codes or Bluetooth from the electric scooter. In the process of activating electric scooters, people should sign up to update their basic personal information, in case of stealing and damaging it, and payment method. Once they arrive at their destinations and end the ride, the app will automatically calculate the distance from where the scooter is activated to the customer’s destination and set an amount of fee at price/distance. Apparently, this idea is stolen from bicycle rental service and other businesses like Uber. However, I think it would be interesting to collect data and calculate to compare whether people use this service for over short distances more than over long distances, and what time of day or night would bring in the most revenue.  Unlike bicycle rental service, this business is using trendy and high-tech transportation, and these can be rented by hours at low cost compared to Uber. Also, people who reluctant using uber due to the risk of crime can use this service and ride to their destinations conveniently and safely in terms of crime risk.

A day to day interaction at this business would be people finding electric scooters around them through an app and riding until their destinations conveniently and safely in terms of crime risk. Our database would work in the background to collect information about each ride (distance, time, price, etc.) The goal of this business is to help people use transportation conveniently and favorably at a low cost.

The tables would include:


1. Time of ride (Real-Time)

2. Price of battery

3. Set price earned by the app

4. Price depending on the weight carried by electric scooter. 

5. Cost per distance (ex: cost per 0.5 mile, 1 mile, etc)
